```python
import pandas as pd
rent= pd.read_csv('Rental.csv', index_col=0)
print(rent.shape)
rent.head()
```

```python
rent.columns
```

```python
#checking null values
rent.isnull().sum()
```

```python
# Dropping NaNs for now in the interest of fast-tracking
# an endpoint for FE team
rent = rent.dropna()

print(rent.shape)
rent.head()
```

```python
rent.dtypes
```

```python
# Can check if dtype is float, and if so, change its
# values to integers

for col in rent.columns.to_list():
    if rent[col].dtype == float:
        rent[col] = [int(x) for x in rent[col].to_list()]
```

```python
rent["Location_Type"].nunique() #> Don't need this one
```

```python
rent.columns
```

```python
rent =rent.drop('Location_Type', axis =1)
print(rent.shape)
rent.head()
```

```python
# Also want to remove the state from the city names
cities = []

for x in rent["Location"]:
    city = x[:-2]
    city = city.replace(",","")
    city = city.strip()
    
    cities.append(city)

rent["city"] = cities

print(rent.shape)
rent.head()
```
